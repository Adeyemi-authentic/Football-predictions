```python
X=new_data.drop(['date','result','opponent'],axis=1)
```

```python
y=new_data['result']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
 from sklearn.metrics import accuracy_score,classification_report
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
rf=RandomForestClassifier()
```

```python

# Define the parameter grid
param_grid = {
    "n_estimators": [10, 50, 70, 100],
    "min_samples_split": [5, 10, 15, 20],
    "criterion": ['gini', 'entropy'],
    "random_state": [1, 5, 10, 15]
}

# Initialize GridSearchCV
rlf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, return_train_score=False)

# Fit the model
rlf.fit(X_train, y_train)
```

```python
rlf.cv_results_
```

```python
pd.DataFrame(rlf.cv_results_)[['param_criterion','param_min_samples_split','param_n_estimators','param_random_state','mean_test_score']]
```

```python
model=RandomForestClassifier(criterion='entropy',n_estimators=10,min_samples_split=15,random_state=5)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
acc=accuracy_score(y_pred,y_test)
print('accuracy of test_data:',acc)
```

```python
# accuracy of train _data
train_pred=model.predict(X_train)
accuracy=accuracy_score(y_train,train_pred)

print('accuracy:',accuracy)
```

```python
print(classification_report(y_test,y_pred))
```
